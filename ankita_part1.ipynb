```python
Report = pd.read_csv("Assignment Data1 - Assignment Data1.csv", low_memory = False)
Report['Call Time']  =pd.to_datetime(Report['Call Time']).dt.date
Y = pd.pivot_table(Report, values = 'Call Time', index = "User Name", aggfunc = pd.Series.nunique, sort = False)
WorkingDays = Y.rename(columns = {'Call Time': 'Working Days'} )
WorkingDays
```

```python
Report = pd.read_csv("Assignment Data1 - Assignment Data1.csv", low_memory = False)
TotalAttempts = Report['User Name'].value_counts(sort = False)
TotalAttempts
```

```python
Report = pd.read_csv("Assignment Data1 - Assignment Data1.csv", low_memory = False)

Report = Report[Report['System Disposition']== "CONNECTED"]
Y = pd.pivot_table(Report, values = 'System Disposition', index = "User Name", aggfunc = 'count')
TotalConnects = Y.rename(columns = {'System Disposition': 'Total Connects'} )
TotalConnects
```

```python
Report = pd.read_csv("Assignment Data1 - Assignment Data1.csv", low_memory = False)

Report = Report[Report['Disposition Class']== "INTERESTED"]
Y = pd.pivot_table(Report, values = 'Disposition Class', index = "User Name", aggfunc = 'count')
TotalInterested = Y.rename(columns = {'Disposition Class': 'Total Interested'} )
TotalInterested
```

```python
Report = pd.read_csv("Assignment Data1 - Assignment Data1.csv", low_memory = False)
X = Report[(Report['Disposition Code']== "Policy Issued") + (Report['Disposition Code']== "Policy_Issued")]
Y = pd.pivot_table(X, values = 'Disposition Code', index = "User Name", aggfunc = 'count')
TotalPolicyIssued = Y.rename(columns = {'Disposition Code': 'Total Policies Issued'} )
TotalPolicyIssued
```

```python
Report = pd.read_csv("Assignment Data1 - Assignment Data1.csv", low_memory = False)
Report['Call Time']  =pd.to_datetime(Report['Call Time']).dt.date

WorkingDays = pd.pivot_table(Report, values = 'Call Time', index = "User Name", aggfunc = pd.Series.nunique, sort = False)
TotalAttempts= Report['User Name'].value_counts(sort = False)
New = pd.concat([WorkingDays, TotalAttempts], axis  = 1)
New
```

```python
Report  = New.rename(columns = {'Call Time':'Working Days', 'User Name': 'Total Attempts'})
Report
```

```python
Report= pd.concat([WorkingDays, TotalAttempts,TotalConnects,TotalInterested, TotalPolicyIssued], axis  = 1)
Report  =  Report.rename(columns = {'User Name': 'Total Attempts'})
```

```python
Report['Average Attempts']  =  Report['Total Attempts']/Report['Working Days']
Report['Average Connects']  =  Report['Total Connects']/Report['Working Days']
Report['Average Interested']  =  Report['Total Interested']/Report['Working Days']
Report['Average Policy Issued']  =  Report['Total Policies Issued']/Report['Working Days']
Report['Connect to Attempt Ratio %']  = Report['Total Connects']/Report['Total Attempts']
Report['Connect to Interested Ratio %']  =  Report['Total Connects']/Report['Total Interested']
Report['Policies Issued to Connect Ratio %']  = Report['Total Policies Issued']/Report['Total Connects']

```
